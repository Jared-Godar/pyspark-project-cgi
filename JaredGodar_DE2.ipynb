##### 4. Write a function to drop columns with more than 1% missing values

```python
percent = 0.01
def drop_null_columns(df, percent):
    to_drop = []
    for col in df.columns:
        if len(df.filter(df[col].isNull()).collect()) / df.count() > percent:
            to_drop.append(col)
    df = df.drop(*to_drop)
    return df, to_drop

sparkDF, dropped_columns = drop_null_columns(sparkDF, percent)
```
